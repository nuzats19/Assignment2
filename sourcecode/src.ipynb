```python
blood_pressure = clean_df['bloodpressure']
```

```python
n_samples = 500
sample_size = 150
boot_means = np.zeros(n_samples)
boot_stds = np.zeros(n_samples)
boot_percentiles = np.zeros(n_samples)
for i in range(n_samples):
    sample = np.random.choice(blood_pressure, size=sample_size, replace=True)
    boot_means[i] = np.mean(sample)
    boot_stds[i] = np.std(sample)
    boot_percentiles[i] = np.percentile(sample, 95)
```

```python
pop_mean = np.mean(blood_pressure)
pop_std = np.std(blood_pressure)
pop_percentile = np.percentile(blood_pressure, 95)
```

```python
fig, axs = plt.subplots(3, figsize=(10, 12))
axs[0].hist(boot_means, bins=20, alpha=0.5, color='blue')
axs[0].axvline(x=pop_mean, color='r', linestyle='--')
axs[0].set_xlabel('Mean BloodPressure')
axs[0].set_ylabel('Frequency')
axs[0].legend(['Population Mean', 'Bootstrap Means'])

axs[1].hist(boot_stds, bins=20,alpha=0.4, color='purple')
axs[1].axvline(x=pop_std, color='r', linestyle='--')
axs[1].set_xlabel('Standard Deviation of BloodPressure')
axs[1].set_ylabel('Frequency')
axs[1].legend(['Population Standard Deviation', 'Bootstrap Standard Deviations'])

axs[2].hist(boot_percentiles, bins=20,alpha=0.5, color='green')
axs[2].axvline(x=pop_percentile, color='r', linestyle='--')
axs[2].set_xlabel('95th Percentile of BloodPressure')
axs[2].set_ylabel('Frequency')
axs[2].legend(['Population 95th Percentile', 'Bootstrap 95th Percentiles'])

plt.savefig("../result/Bootstrapcompbar.jpg")
plt.show()
```
